```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
#Reading the data file. 
data = pd.read_csv("Sortable_FarmData.csv", index_col=0) 
```

## Data_by_People

```python
sns.swarmplot(x='Employee Name', y='Apples Picked', data=data) 
```

### Findings as per the data_by_people graph above. 
  - Alex picked the highest number of apples followed by Jesse. 
  - Riley picked the least number of apples.

## Data_By_Day

```python
#Pivot 1- For every day, the people who worked and number of apples picked.
data_by_day = pd.pivot_table(data, index= 'Day', columns='Employee Name', values="Apples Picked", fill_value=0).plot(kind= "bar", figsize=(9,6), stacked= True)

data_by_day.set(xlabel="Day", ylabel="Apples Picked")
```

### Findings as per the data_by_day graph above. 
  - Day 8th saw the lowest productivity (productivity = the number of apples picked)
  - Day 6th saw the highest productivity.
  - Sam worked for 8 days. Everybody else worked 10 days.

## Data_By_Field

```python
#Pivot 2- For every field, the people who worked and number of apples picked.
data_by_field = pd.pivot_table(data, index= 'Field Assigned', columns='Employee Name', values="Apples Picked", fill_value=0)
sns.heatmap(data_by_field)
```

### Findings as per the data_by_field graph above.

- Alex and Jesse never worked on field C.
- Riley never worked on field B.
- Sam never worked on field A.
- Alex picked the maximum number of apples followed by Jesse.
